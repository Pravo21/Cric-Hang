```python
import tkinter as tk
import random as r
from tkinter import messagebox as m
import pandas as pd
box=tk.Tk(className='Hangman') #Create the GUI with heading

while True:   #Gets the input from dataset with involving any special characters
    df=pd.read_csv('Cricketers.csv')
    inp=r.choice(df['NAME']).strip() #gets the input and removes the junk spaces if any
    if '`' not in inp: #checks any special characters/junk character involved
        break

    
valid_found=0 #stores the number of valid letters found by the user
wrgn=0 #maintains count of wrong answers
wrong_alpha=[]

def press(clicked_alpha,k,num): #Gets invovked and fills the answer boxes when valid alphabet has been selected
    global valid_found
    global wrgn
    old_valid_num=valid_found #assigns the number of valid alphabets to a variable
    for j in range(len(inp)): #iterates through the length and checks whether the selected alphabets matches with inp
        if inp[j].upper()==clicked_alpha:
            if len(k[j].get())==0: #checks whether any letter is already present in that position of clicked alpha
                k[j].insert(0,inp[j].upper()) 
                valid_found+=1 #increments when each valid alphabet is found
    else:
        wrong_alpha.append(clicked_alpha)

    if old_valid_num==valid_found: #checks whether valid alphabet number gets increases, if not chances gets decreased
        if wrong_alpha.count(clicked_alpha)==1:
            wrgn+=1
        if wrgn==3: #Shows warning when 3 wrong attempts given
            m.showwarning('Warning','You have only 2 chances left')
        elif wrgn==5: #Ends the game when all chances have been used and displays the correct answer
            msg='You lost the game, the correct ans is %s'%(inp)
            m.showerror('Game Over',msg)
            box.destroy()
            
    if num==valid_found: #when valid alphabet count matches the total number of characters, user wins
        m.showinfo('Result','You won')
        box.destroy()
def start(): #This function will set the input entry boxes
    
    a=2   #position of the column number for first 13 Alphabet buttons
    b=2   #position of the column number for first 13 Alphabet buttons
    k=[]  #list for storing the variables for input buttons

    wel='Hey %s, welcome to the game'%(j1.get()) #Welcome message to the user
    tk.Label(box,text=wel).grid(row=3) #displays the welcome message
    for i in range(65,91): #creates button and invoke func press when a variable input is clicked or selected
        if a<15: #Creates buttons for getting alphabet inputs for first 13
            tk.Button(box,text=chr(i),command=lambda i=i:press(chr(i),k,num)).grid(row=4,column=a)
            a+=1
        else: #Creates buttons for getting alphabet inputs for next 13 to have a good look of equally spacing boxes
            tk.Button(box,text=chr(i),command=lambda i=i:press(chr(i),k,num)).grid(row=5,column=b)
            b+=1
    
    a=0
    num=0 #Variable for counting number of characters in input excluding spaces
    
    global inp
    for i in range(0,len(inp)): #creates the Entry box for the input 
        if inp[i]!=' ': #create the variables dynamically based on the size of input 
            a+=1
            k.append('e'+str(i)) 
            num+=1
        else: #appends space for empty iterables in the input
            k.append(' ')

    for i in range(len(k)):
        #creates the Entry boxes dynamically based on size of input and assigns it to the variables created in
        #previous step
        if k[i]!=' ': #if there is a space in the input no entry box will be created
            k[i]=tk.Entry(box,width=3)
            k[i].grid(row=6,column=i+2)
            
def chck(): #Checks whether given name is empty or valid, If name is valid invokes the func start
    if len(j1.get())==0:
        m.showerror('Error','Please Enter you Name and then continue')
    else:
        start()
    
tk.Label(text="Find the Cricketer!!!!").grid(row=0,column=0) #First line of welcome text
tk.Label(box,text='Name',font=('Times New Roman',10)).grid(row=1,column=0) #Box for Name Label
j1=tk.Entry(box,fg='red',bg='white') #Entry box for getting the name
j1.grid(row=1,column=1) #position of the name entry box
tk.Button(box,text='start',command=chck).grid(row=2,column=0) #Button for starting the game,invokes the function chck
box.mainloop()
```
